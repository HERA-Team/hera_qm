```python
%matplotlib notebook
from hera_qm import utils
from hera_qm.xrfi import detrend_medfilt
import numpy as np
import matplotlib.pyplot as plt
```

```python
nt = 60
nf = 1024
nbl = 100
rfi_amp = 30
```

```python
np.random.seed(100)
indata = np.random.normal(size=(nt, nf, nbl)) + 1j * np.random.normal(size=(nt, nf, nbl))
indata[45, 300, :] = np.random.normal(scale=rfi_amp, size=nbl) + 1j * np.random.normal(scale=rfi_amp, size=nbl)
for i in range(nbl):
    indata[:, :, i] = detrend_medfilt(np.abs(indata[:, :, i]))
```

```python
indata.imag.min()
```

```python
indata = indata.astype(np.float)
out1 = utils.quadmean(indata, axis=2)
out2 = utils.squadmean(indata, axis=2)
out1 = detrend_medfilt(out1)
out2 = detrend_medfilt(out2)
```

```python
plt.figure()
ax1 = plt.subplot(211)
plt.imshow(out1, aspect='auto')
plt.colorbar()
plt.subplot(212, sharex=ax1, sharey=ax1)
plt.imshow(out2, aspect='auto')
plt.colorbar()
```

```python
plt.figure()
plt.hist(out1.flatten(), alpha=0.5, label='quadmean')
plt.hist(out2.flatten(), alpha=0.5, label='squadmean')
```
